```python
import matplotlib.pyplot as plt
import os
import pandas as pd
import math
```

```python
csvpath = ("oil_sentiments_vs_prices.csv")
df = pd.read_csv(csvpath)

df.head()
```

## Select A Ticker  

```python
ticker = 'MPC'
stock_df = df[df['Ticker'] == ticker]
stock_df
```

```python
Date = stock_df["Date"]
stock_price = stock_df["Price"]
t_polarity = stock_df["Tweet Polarity"]


plt.figure(figsize=(12,8))
plt.scatter(stock_price,t_polarity,s=10*stock_price, c="gold", 
            edgecolor="black", linewidths=1, marker="o", 
            alpha=0.8)
price_min = math.floor(min(stock_price))
price_max = math.ceil(max(stock_price))
plt.title("Tweet Polarity vs. Stock Price",fontsize=18)
plt.ylabel("T-Polarity",fontsize=18)
plt.xlabel("Stock Price",fontsize=18)
plt.xlim(price_min,price_max)
plt.grid(True, color="black", linestyle= '--')

plt.show()
```

## Functions

```python
# Scatter Plot for Ticker
def plot_scatter(ticker):
    stock_df = df[df['Ticker'] == ticker]
    Date = stock_df["Date"]
    stock_price = stock_df["Price"]
    t_polarity = stock_df["Tweet Polarity"]


    plt.figure(figsize=(12,8))
    plt.scatter(stock_price,t_polarity,s=10*stock_price, c="gold", 
                edgecolor="black", linewidths=1, marker="o", 
                alpha=0.8)
    price_min = math.floor(min(stock_price))
    price_max = math.ceil(max(stock_price))
    plt.title("Tweet Polarity vs. Stock Price",fontsize=18)
    plt.ylabel("T-Polarity",fontsize=18)
    plt.xlabel("Stock Price",fontsize=18)
    plt.xlim(price_min,price_max)
    plt.grid(True,color="black", linestyle= '--')

    plt.show()
```

```python
# R-Square for a Ticker
def r_square(ticker):
    x = stock_price
    y = t_polarity
    slope, intercept, r_value, p_value, std_err = stats.linregress(x,y)
    print ("%s r-squared: " % ticker, r_value**2)
    return r_value**2
```

```python
ticker_list = df['Ticker'].unique()
for ticker in ticker_list:
    plot_scatter(ticker)
    r_square(ticker)
```
